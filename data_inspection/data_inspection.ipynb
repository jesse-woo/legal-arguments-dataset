```python
json_file_path = "fewshot-llama-70b_few_shot_generations.json"
summarization_path = "examples.json"
completion_path = "examples-argument-completion.json"
```

```python
import pandas as pd
import json

# Load the JSON data into a DataFrame
try:
    df = pd.read_json(json_file_path)
    
    # Print the data in the rows
    print("Data in rows:")
    for index, row in df.iterrows():
        if index < 10:
            print(f"Row {index + 1}:")
            print(row['text'])
            print("Reference")
            print(row["reference"])
            print(row["prediction"])
            print()
except FileNotFoundError:
    print(f"File not found: {json_file_path}")
except ValueError as e:
    print(f"Error parsing JSON: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
try:
    df_sum = pd.read_json(summarization_path)
    
    # Print the data in the rows
    print("Data in rows:")
    for index, row in df_sum.iterrows():
        if index < 10:
            for key, val in row.items():
                print(f"Row {index + 1}:")
                print(key)
                print(val)
                print()
except FileNotFoundError:
    print(f"File not found: {json_file_path}")
except ValueError as e:
    print(f"Error parsing JSON: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
